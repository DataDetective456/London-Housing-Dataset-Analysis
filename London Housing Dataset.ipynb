# Let us fill the NaN values for each region with the average of the already existing values of each particular area under the "no_of_crimes" column

```python
# Define a function to replace NaN values with the area's average
def fill_missing_crimes(row, average_crimes):
    if pd.isna(row['no_of_crimes']):
        return average_crimes[row['area']]
    else:
        return row['no_of_crimes']

# Apply the function to fill missing values in the 'no_of_crimes' column
data['no_of_crimes'] = data.apply(lambda row: fill_missing_crimes(row, average_crimes), axis=1)

# Display the updated DataFrame
print(data)
```

```python
# Now, instead of dropping the null values, Let us fill it with randomnumbers generated from the non-null rows of the "no_of_crimes" colum
```

```python
# Extract non-null values from the 'no_of_crimes' column
non_null_crimes = data['no_of_crimes'].dropna().values

# Function to fill NaN values with random choice from existing non-null values
def fill_with_random_crimes(row):
    if pd.isna(row['no_of_crimes']):
        return np.random.choice(non_null_crimes)
    else:
        return row['no_of_crimes']

# Apply the function to the DataFrame
data['no_of_crimes'] = data.apply(fill_with_random_crimes, axis=1)

# Display the updated DataFrame
print(data)
```
